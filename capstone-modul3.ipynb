### ***Matrix Evaluation***

1. **Mean Absolute Error (MAE):** MAE measures the average absolute difference between actual and predicted values. A lower MAE indicates better model performance. We use MAE because the target variable, Price, has outliers. MAE is less sensitive to outliers as it calculates the absolute value of errors, ensuring outliers with large errors do not disproportionately impact the metric.

2. **Mean Absolute Percentage Error (MAPE):** MAPE measures the average percentage difference between actual and predicted values. A lower MAPE signifies better accuracy. Similar to MAE, MAPE is less sensitive to outliers as it calculates relative error percentages, ensuring outliers do not heavily skew the results.

3. **Root Mean Squared Error (RMSE):** RMSE measures the square root of the average squared differences between actual and predicted values. RMSE penalizes larger errors more than smaller ones, providing a clear indication of model performance in absolute terms. While sensitive to outliers, it is useful for understanding the magnitude of prediction errors.

### ***Interquartile Range***

```python
# Dictionary to store outliers' information
outliers_info = {}

# List of numerical columns
numerical_columns = ['Year', 'Engine_Size', 'Mileage', 'Price']

for column in numerical_columns:
    # Calculate Q1, Q3, and IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Calculate bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers_condition = (df[column] < lower_bound) | (df[column] > upper_bound)
    outliers = df[outliers_condition]

    # Store information in the dictionary
    outliers_info[column] = {
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'number_of_outliers': outliers_condition.sum(),  # Count of outliers
        'outliers_indices': outliers.index.tolist()  # Indices of outliers
    }


# summary for outliers_summary

outliers_summary = {key: {'bounds': (value['lower_bound'], value['upper_bound']), 
                          'number_of_outliers': value['number_of_outliers']} 
                    for key, value in outliers_info.items()}

# Creating DataFrame from 'outliers_summary'
outliers_summary_df = pd.DataFrame.from_dict(outliers_summary, orient='index')

# Splitting 'bounds' into 'lower_bound' and 'upper_bound' columns
outliers_summary_df[['lower_bound', 'upper_bound']] = outliers_summary_df['bounds'].apply(pd.Series)

# Dropping the original 'bounds' column
outliers_summary_df.drop('bounds', axis=1, inplace=True)

# Reordering the columns for a better view
outliers_summary_df = outliers_summary_df[['lower_bound', 'upper_bound', 'number_of_outliers']]

# Display DataFrame as a styled table
styled_table = outliers_summary_df.style.format('{:.0f}')
styled_table

```

### ***Matrix Explanation***

**Root Mean Square Error (RMSE):**
- Advantages: It is sensitive to larger errors and can reveal how tightly the observed data 
clusters around the predicted values1. It’s also easy to interpret since the units of 
measurement match the target variable.
- Disadvantages: It is sensitive to outliers and can exaggerate results if there are 
outliers in the dataset.

**Mean Absolute Error (MAE):**
- Advantages: It’s simple to understand and calculate. The result will have the same unit 
as that of the output.
- Disadvantages: It’s not differentiable at all points, which can be a problem for certain 
optimization algorithms.

**Mean Absolute Percentage Error (MAPE):**
- Advantages: It’s easy to understand as the error is a percentage. It can be more 
beneficial because it shows the percentage difference.
- Disadvantages: It’s not possible to use when actual values can be at or close to zero. 
It’s also asymmetric, meaning that it reports higher errors if the predicted value is 
higher than the actual value and reports lower errors if the predicted value is lower than 
the actual.

This phase optimizes the model’s performance by f**ine-tuning hyperparameters**. It involves systematically adjusting algorithm parameters to achieve the best combination for **model accuracy and generalization**. Techniques such as **grid search or random search** are employed to find the optimal settings that improve model performance.

```python
hyperparam_space_catboost = {
    'model__random_strength': [0.3, 0.5, 0.7, 1.0],  # Increased values
    'model__max_bin': list(range(2, 41, 2)),
    'model__bootstrap_type': ['Bayesian', 'Bernoulli', 'MVS', 'No'],
    'model__min_data_in_leaf': list(range(5, 31, 5)),  # Increased values
    'model__max_ctr_complexity': [1, 2, 3],  # Decreased range
    'model__learning_rate': [0.001, 0.01, 0.1],  # Reduced options
    'model__iterations': list(range(100, 401, 50)),
    'model__depth': list(range(5, 13)),
    'model__l2_leaf_reg': list(range(3, 9)),  # Increased lower bound
}
```

**Parameter Explanations and Adjustments for CatBoost:**

1. **random_strength**: Balances categorical feature influence during splits. Tune this to enhance performance with categorical features.
2. **max_bin**: Determines the number of bins for feature values. Too small can underfit; too large can overfit.
3. **bootstrap_type**: Defines the sampling method. Different methods affect performance.
4. **min_data_in_leaf**: Prevents overfitting by setting a minimum number of samples per leaf. Higher values prevent overfitting but may underfit.
5. **max_ctr_complexity**: Limits the number of combined categorical features. Higher values capture complex patterns but can overfit.
6. **learning_rate**: Controls step size in iterations. Lower values improve accuracy but slow down learning.
7. **iterations**: Number of trees built. More trees capture complexity but can overfit.
8. **depth**: Tree depth. Deeper trees capture complexity but may overfit.
9. **l2_leaf_reg**: Controls L2 regularization to prevent overfitting.

**Conclusion:**

To address overfitting in CatBoost:
- Increase **random_strength**.
- Increase **min_data_in_leaf**.
- Decrease **max_ctr_complexity**.
- Increase **l2_leaf_reg**.
